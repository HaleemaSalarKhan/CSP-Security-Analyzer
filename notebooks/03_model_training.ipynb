# 03 - Model Training

This notebook trains a `RandomForestClassifier` using an 80/20 train-test split.

```python
from pathlib import Path
import sys
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(PROJECT_ROOT))

from src.model_utils import train_random_forest

DATA_DIR = PROJECT_ROOT / 'data'
MODEL_DIR = PROJECT_ROOT / 'models'
```

```python
df = pd.read_csv(DATA_DIR / 'csp_features_labeled.csv')
df['label'].value_counts()
```

```python
model, X_train, X_test, y_train, y_test = train_random_forest(
    df,
    MODEL_DIR / 'csp_random_forest.pkl',
    random_state=42,
)

predictions = model.predict(X_test)
accuracy_score(y_test, predictions)
```

```python
print(classification_report(y_test, predictions))
```

```python
confusion_matrix(y_test, predictions, labels=['Weak', 'Medium', 'Strong'])
```
